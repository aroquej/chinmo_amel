### **Fetching gene information data for genes quantified by Sleuth from NCBI.**

#### Herb et al. (2012)

```python
import pandas as pd
from Bio import Entrez
from time import sleep
import re 

#NCBI API Key
Entrez.email = "a.roque0@usp.br"

#Function to fetch the information from NCBI
def fetch_ncbi_info(transcript_id):
    try:
        print(f"Consulting: {transcript_id}")

        # Access the NCBI database to get the information
        handle = Entrez.efetch(db="nucleotide", id=transcript_id, rettype="gb", retmode="text")
        record = handle.read()
        handle.close()

        # Extract the gene name
        gene_name_match = re.search(r'/gene="([^"]+)"', record)
        gene_name = gene_name_match.group(1) if gene_name_match else None

        # Extract the BeeBase ID (ex: GB413947)
        beebase_match = re.search(r'/db_xref="BEEBASE:(GB\d+)"', record)
        gene_beebase = beebase_match.group(1) if beebase_match else None

        # Extract the GeneID (ex: 100287)
        gene_id_match = re.search(r'/db_xref="GeneID:(\d+)"', record)
        gene_id = gene_id_match.group(1) if gene_id_match else None


        # Extracts the product name
        product_match = re.search(r"DEFINITION\s+(.*)", record)
        product = product_match.group(1) if product_match else None

        return pd.Series([gene_name, gene_beebase, gene_id, product])

    except Exception as e:
        print(f"Error while searching {transcript_id}: {e}")
        return pd.Series([None, None, None, None, None])

# Load the table with the results
df = pd.read_csv(r"..\data\sleuth_results_Herb.csv", encoding='latin1')


# Add new columns to the table
df[['gene_name', 'gene_beebase', 'gene_id', 'product']] = df['target_id'].apply(
    lambda x: fetch_ncbi_info(x)
)

# Save the table with the new columns as a CSV file
df.to_csv(r"..\02_enriched_tables\sleuth_results_Herb_enriched.csv", index=False)

print("Completed table saved.")
```

### **Identifying differentially expressed genes that present Chinmo's motif in their 1 Kb upstream promoter through intersection with the table 'all_promoters_with_motifs_chinmo' obtained through FIMO motif scan on MEME suite environment.**

#### Khamis et al. (2015)

```python
import pandas as pd
from matplotlib_venn import venn2
from matplotlib import pyplot as plt
import os
# Path to output directory
output_dir = r"03_genes_with_motifs"
output_file = os.path.join(output_dir, "genes_with_motif_Khamis.csv")

# Loading the Sleuth table after information recovery.
df_DEG_khamis = pd.read_csv(r"02_enriched_tables\sleuth_results_Khamis_enriched.csv", encoding='latin1')

# Intersection limited to differentially expressed genes. q-val set as <=0.05 and 0.5 <= |b| >= 0.5. b corresponds to the log2 fold change.
df_DEG_khamis = df_DEG_khamis[(df_DEG_khamis['qval'] <= 0.05) & ((df_DEG_khamis['b'] <= -0.5) | (df_DEG_khamis['b'] >= 0.5))]

# Creating a dictornary of gene names.
genes_DEG = set(df_DEG_khamis['gene_name'].dropna())

# Loading table with all genes from the genome that present Chinmo's motif
df_promoters = pd.read_csv(r"data\all_promoters_with_motifs_chinmo.csv", encoding='latin1')
genes_promoters = set(df_promoters['gene_names'].dropna())

# Performing intersection between dictionaries.
common_genes = genes_DEG.intersection(genes_promoters)
df_common_genes = df_promoters[df_promoters['gene_names'].isin(common_genes)]
df_common_genes = df_common_genes.merge(df_DEG_khamis[['gene_name', 'gene_id']], left_on='gene_names', right_on='gene_name', how='left')
df_common_genes = df_common_genes[['gffid','gene_id', 'gene_names', 'gene_beebase', 'product']]
df_common_genes = df_common_genes.drop_duplicates(subset=['gene_id'])
motif_counts = df_promoters[df_promoters['gene_names'].isin(common_genes)].groupby('gene_names')['gffid'].count()
df_common_genes['motif_count'] = df_common_genes['gene_names'].map(motif_counts)
# Saving new table as CSV.
df_common_genes.to_csv(output_file, index=False)

print(f"Table with {len(common_genes)} common genes saved in: {output_file}")

# Creating a Venn's diagram
plt.figure(figsize=(8, 6))
venn = venn2([genes_DEG, genes_promoters], ('Table Khamis et al. 2015', 'Promotores'))
plt.title("Venn's diagram - DEG's with Chinmo's motif")
plt.show()
```

### **Identifying overlapping DEGs between the studies.**

```python
import os
import pandas as pd
from matplotlib_venn import venn3
from matplotlib import pyplot as plt

output_dir = r"03_genes_with_motifs"
output_file = os.path.join(output_dir, "common_genes_all.csv")


# Loading tables of co-differentially expressed genes with CHINMO's motif
df_justino = pd.read_csv(r"03_genes_with_motifs\genes_with_motif_Justino_JH_vs_AC.csv", encoding='latin1')
df_khamis = pd.read_csv(r"03_genes_with_motifs\genes_with_motif_Khamis.csv", encoding='latin1')
df_warner = pd.read_csv(r"03_genes_with_motifs\genes_with_motif_Warner_L5.csv", encoding='latin1')

# Extracting gene_names
genes_justino = set(df_justino['gene_names'].dropna().unique())
genes_khamis = set(df_khamis['gene_names'].dropna().unique())
genes_warner = set(df_warner['gene_names'].dropna().unique())

# Creating a Venn's diagram.
plt.figure(figsize=(10, 8))
venn = venn3([genes_justino, genes_khamis, genes_warner], 
             ('This study', 'Khamis et al. (2015)', 'Warner et al. (2019)'),
             set_colors=("#FFCC00", "#CC9900", "#0000FF"))
plt.title("Shared DEGs with Chinmo's motif")
plt.show()

# Creating a dataframe of the common DEGs between the three transcriptomes.
common_genes_all = genes_justino.intersection(genes_khamis, genes_warner)
df_common_genes_all = df_justino[df_justino['gene_names'].isin(common_genes_all)]

# Saving the dataframe as a csv file.

common_output_file = os.path.join(output_dir, "common_genes_all.csv")
df_common_genes_all.to_csv(common_output_file, index=False)
print(f"Table with {len(df_common_genes_all)} common genes saved in: {output_file}")

```
